```python
#Carga de librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as mode
```

```python
#Carga de archivo CSV
df= pd.read_csv("listingsba.csv")
df
```

```python
df0=df.drop(['bathrooms_text','neighbourhood_group_cleansed'], axis=1)
data=df0.drop(['calendar_updated','calculated_host_listings_count_shared_rooms','calculated_host_listings_count_private_rooms','calculated_host_listings_count_entire_homes','calculated_host_listings_count','number_of_reviews_l30d',"calendar_last_scraped",'minimum_minimum_nights','maximum_minimum_nights', 'minimum_maximum_nights','maximum_maximum_nights','minimum_nights_avg_ntm', 'maximum_nights_avg_ntm','has_availability','availability_30','availability_60','availability_90','availability_365'], axis=1)
data
```

```python
#Identificación de valores nulos
#Total de nulos 114820
null=data.isnull().sum() .sum()
null
```

```python
#Sustitución con diferentes métodos 

data["name"]= data["name"].fillna('Desconocido')
data["description"]= data["description"].fillna('Se desconoce la descripción del lugar')
data["host_response_time"]= data["host_response_time"].fillna('El anfitrión responderá pronto')
data["host_response_rate"]= data["host_response_rate"].fillna(method='ffill')
data["host_acceptance_rate"]= data["host_acceptance_rate"].fillna(method='ffill')
data["host_acceptance_rate"]= data["host_acceptance_rate"].fillna(method='bfill')
data["neighbourhood"]= data["neighbourhood"].fillna(method='ffill')
data['bathrooms']=data['bathrooms'].fillna(1)
data['bedrooms']=data['bedrooms'].fillna(1)
data['beds']=data['beds'].fillna(1)
data["first_review"]=data["first_review"].fillna('00/00/0000')
data["last_review"]=data["last_review"].fillna('00/00/0000')
data["review_scores_rating"]=data["review_scores_rating"].fillna('Aún no es posible')
data["neighborhood_overview"]=data["neighborhood_overview"].fillna('Aún no es posible')
data["review_scores_accuracy"]=data["review_scores_accuracy"].fillna('Aún no es posible')

data["review_scores_cleanliness"]=data["review_scores_cleanliness"].fillna('Aún no es posible')
data["review_scores_checkin"]=data["review_scores_checkin"].fillna('Aún no es posible')
data["review_scores_communication"]=data["review_scores_communication"].fillna('Aún no es posible')
data["review_scores_location"]=data["review_scores_location"].fillna('Aún no es posible')
data["host_neighbourhood"]=data["host_neighbourhood"].fillna('No disponible')
data["review_scores_value"]=data["review_scores_value"].fillna('Aún no es posible')
data["license"]=data["license"].fillna('No aplica')

data['reviews_per_month']=data['reviews_per_month'].fillna(0)
data["license"]= data["license"].fillna('Información no disponible') 
data["host_is_superhost"]= data["host_is_superhost"].fillna('Se desconoce') 
data["host_location"]= data["host_location"].fillna('Información no disponible') 
data["host_about"]= data["host_about"].fillna('Información no disponible')
```
